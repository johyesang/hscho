```python
#데이터 준비
#도미 길이, 무게 데이터
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]

#빙어 길이, 무게 데이터
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

#산점도 그래프 모듈 추가
import matplotlib.pyplot as plt

#도미의 길이, 무게 산점도 그래프
bream_scatter = plt
bream_scatter.scatter(bream_length, bream_weight)
bream_scatter.xlabel('length')
bream_scatter.ylabel('weight')
bream_scatter.title('Scatter graph of Bream')
bream_scatter.show()

#빙어의 길이, 무게 산점도 그래프
smelt_scatter = plt
smelt_scatter.scatter(smelt_length, smelt_weight)
smelt_scatter.xlabel('length')
smelt_scatter.ylabel('weight')
smelt_scatter.title('Scatter graph of Bream')
smelt_scatter.show()

#도미-빙어 섞은 산점도 그래프
mixed_scatter = plt
mixed_scatter.scatter(bream_length, bream_weight)
mixed_scatter.scatter(smelt_length, smelt_weight)
mixed_scatter.xlabel('length')
mixed_scatter.ylabel('weight')
mixed_scatter.title('Mixed Scatter graph of Bream and Smelt')
mixed_scatter.show()

#데이터 합쳐서 분류 모델 만들기
#KNN 분류 모델
length = bream_length + smelt_length
weight = bream_weight + smelt_weight

fish_data = [[l,w] for l, w in zip(length, weight)]
fish_target = [1]*35+[0]*14 #도미는 1, 빙어는 0 으로 가정함.

# sklearn KNC 불러오기
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier() #기본 k의 수는 5개로 되어 있음
kn.fit(fish_data, fish_target)
kn_score = kn.score(fish_data, fish_target)

kn.predict([[30, 600]]) #학습한 모델로 예측해보기

#하이퍼파라미터 변경 n개 만큼
kn_49 = KNeighborsClassifier(n_neighbors = 49) #기본 k의 수는 5개로 되어 있음
kn_49.fit(fish_data, fish_target)
kn_49_score = kn_49.score(fish_data, fish_target)

print(f'number of neighbors 4  : score {kn_score}\nnumber of neighbors 49 : score {kn_49_score:4f}')
```
